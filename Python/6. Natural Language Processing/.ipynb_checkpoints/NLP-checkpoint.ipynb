```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

table = pd.read_csv('Restaurant_Reviews.tsv', delimiter = '\t', quoting = 3)
table   
```

```python
#Cleaning of text
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

#Stemming all words to remove plurals and removing stop words
def clean_text(table):
    corpus = []
    for i in range(0, 1000):
        review = re.sub('[^a-zA-Z]', ' ', table['Review'][i])
        review = review.lower()
        review = review.split()
        ps = PorterStemmer()
        all_stopwords = stopwords.words('english')
        all_stopwords.remove('not')
        review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
        review = ' '.join(review)
        corpus.append(review)
    return corpus
table['Review'] = clean_text(table)
table
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
y = table.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
model = LogisticRegression(random_state = 0)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print("accuracy of the model is " , accuracy_score(y_pred,y_test)*100, '%')
```

```python
#Custom Input

def custom_input(sentence):
    corpus = []
    review = re.sub('[^a-zA-Z]', ' ', sentence)
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
    X = cv.transform(corpus).toarray()
    return model.predict(X)
review = input("Input review: - ")
custom_input(review)
```
